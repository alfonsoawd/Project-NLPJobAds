```python
# Create a cleaned copy of the original DataFrame
df_clean = df.copy()

# Clean the 'location_departement' column
def clean_departement_column(df, column):
    df = df.copy()  # avoid SettingWithCopyWarning

    # Convert to string and strip spaces
    df[column] = df[column].astype(str).str.strip()

    # Remove ".0" if present
    df[column] = df[column].str.replace(r'\.0$', '', regex=True)

    # Replace known invalid entries with pd.NA
    invalids = ['nan', '', 'NaN', 'None', '0', '00', '<NA>', 'NA']
    df[column] = df[column].where(~df[column].isin(invalids), pd.NA)

    # Pad numbers like '2' to '02', but keep valid string codes like '2A' and '2B' untouched
    df[column] = df[column].apply(
        lambda x: x.zfill(2) if isinstance(x, str) and x.isdigit() and len(x) == 1 else x
    )

    return df

# Apply cleaning to location_departement
df_clean = clean_departement_column(df_clean, 'location_departement')

# Check results
print("Unique values:", sorted(df_clean['location_departement'].dropna().unique()))
print("Missing values:", df_clean['location_departement'].isna().sum())
```

```python
def tabulate_categorical(df_clean, column, sort=True, normalize=True):

    counts = df_clean[column].value_counts(dropna=False, sort=sort)
    total = counts.sum()
    
    tab_df = counts.to_frame(name='Count')
    if normalize:
        tab_df['Percent'] = 100 * tab_df['Count'] / total
        tab_df['Percent'] = tab_df['Percent'].round(2)
    
    print(f"\nVariable: {column}")
    print(f"Unique categories: {df_clean[column].nunique(dropna=False)}")
    display(tab_df)

tabulate_categorical(df_clean, 'famille_pro22')
tabulate_categorical(df_clean, 'location_departement')
```

# 2. Looking how advertised attributes differ between ads that mention salary and those that do not

```python
df_salary_disclosed = df_clean[df_clean["is_salary"] == 1]
df_salary_hidden = df_clean[df_clean["is_salary"] == 0]

agg_attributes = [col for col in df_clean.columns if col.startswith("agg_")]
```

```python
print(f" Job ads with salary disclosed: {df_salary_disclosed.shape[0]}")
print(f" Job ads without salary disclosed: {df_salary_hidden.shape[0]}")
print(f" Total ads used: {df_salary_disclosed.shape[0] + df_salary_hidden.shape[0]}")
```

```python
# Step 1 ‚Äî Compute the share of job ads mentioning each aggregated attribute
# For ads that DO disclose salary
share_disclosed = df_salary_disclosed[agg_attributes].mean().rename("With Salary")

# For ads that DO NOT disclose salary
share_hidden = df_salary_hidden[agg_attributes].mean().rename("Without Salary")

# Step 2 ‚Äî Combine both series into a single DataFrame and compute the difference
comparison_df = pd.concat([share_disclosed, share_hidden], axis=1)
comparison_df["Difference"] = comparison_df["With Salary"] - comparison_df["Without Salary"]

# Step 3 ‚Äî Clean column names and prepare for plotting
comparison_df = comparison_df.reset_index().rename(columns={"index": "Attribute"})
comparison_df["Attribute"] = comparison_df["Attribute"].str.replace("agg_", "")

# Step 4 ‚Äî Run z-test for difference in proportions and store p-values
p_values = []

for col in agg_attributes:
    # Count how many times the attribute is mentioned in each group
    count_salary = df_salary_disclosed[col].sum()
    count_nosalary = df_salary_hidden[col].sum()

    # Total number of observations in each group
    n_salary = df_salary_disclosed.shape[0]
    n_nosalary = df_salary_hidden.shape[0]

    # Run two-proportion z-test
    try:
        stat, pval = proportions_ztest([count_salary, count_nosalary], [n_salary, n_nosalary])
    except:
        pval = 1.0  # If test fails (e.g., only zeros), assume not significant

    p_values.append(pval)

# Step 5 ‚Äî Add p-values to the main comparison DataFrame
comparison_df["p_value"] = p_values

# Step 6 ‚Äî Filter only those attributes with statistically significant differences (e.g., p < 0.05)
significant_agg_df = comparison_df[comparison_df["p_value"] < 0.05].copy()

# Sort by difference for clearer visual comparison
significant_agg_df = significant_agg_df.sort_values(by="Difference", ascending=False)
```

```python
plt.figure(figsize=(8, 12))
sns.barplot(
    data=significant_agg_df, 
    x="Difference", 
    y="Attribute", 
    palette="coolwarm"
)

plt.axvline(0, color='gray', linestyle='--')
#plt.title("Significant Differences in Attribute Prevalence\n(With vs. Without Salary Disclosure)")
plt.xlabel("Difference in Share (With Salary - Without Salary)")
plt.tight_layout()
plt.show()
```

```python
# Step 1 ‚Äî Identify valid subattribute columns (binary, non-aggregated, excluding 'is_salary')
sub_attributes = [
    col for col in df_clean.columns
    if not col.startswith("agg_")                         # exclude aggregated attributes
    and col != "is_salary"                                # exclude salary indicator
    and df_clean[col].dropna().isin([0, 1]).all()        # only keep binary columns
    and df_clean[col].nunique() > 1                      # must vary
]

# Step 2 ‚Äî Compute the share of job ads mentioning each subattribute
# For ads that disclose salary
share_disclosed_sub = df_salary_disclosed[sub_attributes].mean().rename("With Salary")

# For ads that do NOT disclose salary
share_hidden_sub = df_salary_hidden[sub_attributes].mean().rename("Without Salary")

# Step 3 ‚Äî Combine into single DataFrame and compute difference
comparison_df_sub = pd.concat([share_disclosed_sub, share_hidden_sub], axis=1)
comparison_df_sub["Difference"] = comparison_df_sub["With Salary"] - comparison_df_sub["Without Salary"]

# Step 4 ‚Äî Prepare for plotting
comparison_df_sub = comparison_df_sub.reset_index().rename(columns={"index": "Attribute"})

# Step 5 ‚Äî Run z-tests for each subattribute and store p-values
p_values_sub = []

for attr in sub_attributes:
    # Count how many times the attribute is present in each group
    count_salary = df_salary_disclosed[attr].sum()
    count_nosalary = df_salary_hidden[attr].sum()

    # Total observations in each group
    n_salary = df_salary_disclosed.shape[0]
    n_nosalary = df_salary_hidden.shape[0]

    # Run z-test for difference in proportions
    try:
        stat, pval = proportions_ztest([count_salary, count_nosalary], [n_salary, n_nosalary])
    except:
        pval = 1.0  # If the test fails, mark as not significant

    p_values_sub.append(pval)

# Step 6 ‚Äî Add p-values to the DataFrame
comparison_df_sub["p_value"] = p_values_sub

# Step 7 ‚Äî Keep only significant results (e.g., p < 0.05) and sort
significant_sub_df = comparison_df_sub[comparison_df_sub["p_value"] < 0.05].copy()
significant_sub_df = significant_sub_df.sort_values(by="Difference", ascending=False)
```

```python
plt.figure(figsize=(8, 12))
sns.barplot(
    data=significant_sub_df, 
    x="Difference", 
    y="Attribute", 
    palette="coolwarm"
)

plt.axvline(0, color='gray', linestyle='--')
plt.title("Significant Differences in Subattribute Prevalence\n(With vs. Without Salary Disclosure)")
plt.xlabel("Difference in Share (With Salary - Without Salary)")
plt.tight_layout()
plt.show()
```

## Regression to control these differences by industry and location 

```python
# List of target attributes (binary columns that start with "agg_")
agg_attributes = [col for col in df_clean.columns if col.startswith("agg_")]

# Store results in a list
regression_results = []

# Loop over each attribute
for target in agg_attributes:
    print(f"Running regression for: {target}")
    
    # Define variables
    cols = [target, "is_salary", "famille_pro22", "location_departement"]
    df_subset = df_clean[cols].dropna()

    # Collapse to unique combinations to reduce computation
    df_collapsed = df_subset.groupby(cols).size().reset_index(name="weight")

    # Define model formula
    formula = f'Q("{target}") ~ is_salary + C(famille_pro22) + C(location_departement)'

    try:
        # Estimate weighted logistic regression (GLM with binomial family)
        model = smf.glm(
            formula=formula,
            data=df_collapsed,
            family=sm.families.Binomial(),
            freq_weights=df_collapsed["weight"]
        ).fit()

        # Extract coefficient and p-value for is_salary
        coef = model.params.get("is_salary", float('nan'))
        pval = model.pvalues.get("is_salary", float('nan'))

        # Store result
        regression_results.append({
            "Attribute": target.replace("agg_", ""),
            "Coef_is_salary": coef,
            "P_value": pval
        })

    except Exception as e:
        print(f"‚ö†Ô∏è Failed for {target}: {e}")

# Create DataFrame with results
results_df = pd.DataFrame(regression_results)

# Optional: Sort by effect size or significance
results_df = results_df.sort_values(by="Coef_is_salary", ascending=False)

# Show results
import matplotlib.pyplot as plt
import seaborn as sns

# Plot only significant coefficients
significant_df = results_df[results_df["P_value"] < 0.05]

plt.figure(figsize=(10, 6))
sns.barplot(data=significant_df, x="Coef_is_salary", y="Attribute", color="coral")
plt.axvline(0, color='gray', linestyle='--')
plt.xlabel("Effect of Salary Disclosure on Attribute Presence")
#plt.title("Logit Coefficient for Salary Disclosure (Significant Only)")
plt.tight_layout()
plt.show()
```

## For aggregate attributes

```python
# Define the target attribute
target = "agg_Attributs du lieu de travail"

# Define model specifications incrementally (no interaction)
fe_sets = {
    "Industry FE": ["famille_pro22"],
    "Industry + Location FE": ["famille_pro22", "location_departement"]
}

# Dictionary to store McFadden pseudo R¬≤ results
results = {}

for label, fe_vars in fe_sets.items():
    print(f"\nüîπ Running model with: {label}")

    # Select relevant columns
    cols = [target] + fe_vars
    df_subset = df_clean[cols].dropna()

    # Collapse identical rows and count frequency
    df_collapsed = df_subset.groupby(cols).size().reset_index(name="weight")

    # Create formula (no interaction)
    fe_terms = " + ".join([f"C({var})" for var in fe_vars])
    formula = f'Q("{target}") ~ {fe_terms}'

    # Fit GLM (logit) with frequency weights
    model = smf.glm(
        formula=formula,
        data=df_collapsed,
        family=sm.families.Binomial(),
        freq_weights=df_collapsed["weight"]
    ).fit()

    # Compute McFadden's pseudo R¬≤ manually
    ll_model = model.llf
    ll_null = model.null_deviance / -2
    pseudo_r2 = 1 - (ll_model / ll_null)

    # Store and display result
    results[label] = pseudo_r2
    print(f"McFadden's Pseudo R¬≤ for {label}: {pseudo_r2:.4f}")
```

```python
# Define model specifications (incremental, no interaction)
fe_sets = {
    "Industry FE": ["famille_pro22"],
    "Industry + Location FE": ["famille_pro22", "location_departement"]
}

# Lista de atributos que empiezan por 'agg_'
agg_attributes = [col for col in df_clean.columns if col.startswith("agg_")]

# Lista para guardar los resultados
pseudo_r2_results = []

# Loop por cada atributo y cada especificaci√≥n de efectos fijos
for target in agg_attributes:
    for label, fe_vars in fe_sets.items():
        print(f"\nüîπ Running model for {target} with: {label}")
        
        # Subset y limpieza
        cols = [target] + fe_vars
        df_subset = df_clean[cols].dropna()

        # Colapsar duplicados y contar frecuencia
        df_collapsed = df_subset.groupby(cols).size().reset_index(name="weight")

        # Crear f√≥rmula sin interacci√≥n
        fe_terms = " + ".join([f"C({var})" for var in fe_vars])
        formula = f'Q("{target}") ~ {fe_terms}'

        try:
            # Ajustar modelo GLM con pesos de frecuencia
            model = smf.glm(
                formula=formula,
                data=df_collapsed,
                family=sm.families.Binomial(),
                freq_weights=df_collapsed["weight"]
            ).fit()

            # Calcular pseudo R¬≤ de McFadden manualmente
            ll_model = model.llf
            ll_null = model.null_deviance / -2
            pseudo_r2 = 1 - (ll_model / ll_null)

            # Guardar resultado
            pseudo_r2_results.append({
                "Attribute": target.replace("agg_", ""),
                "FE_Set": label,
                "Pseudo_R2": pseudo_r2
            })

        except Exception as e:
            print(f"‚ö†Ô∏è Error with {target} / {label}: {e}")
            pseudo_r2_results.append({
                "Attribute": target.replace("agg_", ""),
                "FE_Set": label,
                "Pseudo_R2": None,
                "Error": str(e)
            })

# Convertir resultados a DataFrame (si lo necesitas luego)
pseudo_r2_df = pd.DataFrame(pseudo_r2_results)
```

```python
# Define model specifications (incremental, no interaction)
fe_sets = {
    "Industry FE": ["famille_pro22"],
    "Industry + Location FE": ["famille_pro22", "location_departement"],
    "Industry + Location FE + wage present FE": ["famille_pro22", "location_departement", "is_salary"]
}

# Lista de atributos que empiezan por 'agg_'
agg_attributes = [col for col in df_clean.columns if col.startswith("agg_")]

# Lista para guardar los resultados
pseudo_r2_b_results = []

# Loop por cada atributo y cada especificaci√≥n de efectos fijos
for target in agg_attributes:
    for label, fe_vars in fe_sets.items():
        print(f"\nüîπ Running model for {target} with: {label}")
        
        # Subset y limpieza
        cols = [target] + fe_vars
        df_subset = df_clean[cols].dropna()

        # Colapsar duplicados y contar frecuencia
        df_collapsed = df_subset.groupby(cols).size().reset_index(name="weight")

        # Crear f√≥rmula sin interacci√≥n
        fe_terms = " + ".join([f"C({var})" for var in fe_vars])
        formula = f'Q("{target}") ~ {fe_terms}'

        try:
            # Ajustar modelo GLM con pesos de frecuencia
            model = smf.glm(
                formula=formula,
                data=df_collapsed,
                family=sm.families.Binomial(),
                freq_weights=df_collapsed["weight"]
            ).fit()

            # Calcular pseudo R¬≤ de McFadden manualmente
            ll_model = model.llf
            ll_null = model.null_deviance / -2
            pseudo_r2 = 1 - (ll_model / ll_null)

            # Guardar resultado
            pseudo_r2_b_results.append({
                "Attribute": target.replace("agg_", ""),
                "FE_Set": label,
                "Pseudo_R2": pseudo_r2
            })

        except Exception as e:
            print(f"‚ö†Ô∏è Error with {target} / {label}: {e}")
            pseudo_r2_b_results.append({
                "Attribute": target.replace("agg_", ""),
                "FE_Set": label,
                "Pseudo_R2": None,
                "Error": str(e)
            })

# Convertir resultados a DataFrame (si lo necesitas luego)
pseudo_r2_b_df = pd.DataFrame(pseudo_r2_b_results)
```

## For subattributes

```python
# Define model specifications (incremental, no interaction)
fe_sets = {
    "Industry FE": ["famille_pro22"],
    "Industry + Location FE": ["famille_pro22", "location_departement"],
    "Industry + Location FE + wage present FE": ["famille_pro22", "location_departement", "is_salary"]
}

# Subattributes list
sub_attributes = [
    col for col in df_clean.columns
    if not col.startswith("agg_")                                # no es categor√≠a agregada
    and col != "is_salary"                                       # excluir is_salary
    and df_clean[col].dropna().isin([0, 1]).all()               # dummy 0/1
    and df_clean[col].nunique() > 1                             # tiene varianza
]


# Lista para guardar los resultados
pseudo_r2_results_sub_b = []

# Loop por cada atributo y cada especificaci√≥n de efectos fijos
for target in sub_attributes:
    for label, fe_vars in fe_sets.items():
        print(f"\nüîπ Running model for {target} with: {label}")
        
        # Subset y limpieza
        cols = [target] + fe_vars
        df_subset = df_clean[cols].dropna()

        # Colapsar duplicados y contar frecuencia
        df_collapsed = df_subset.groupby(cols).size().reset_index(name="weight")

        # Crear f√≥rmula sin interacci√≥n
        fe_terms = " + ".join([f"C({var})" for var in fe_vars])
        formula = f'Q("{target}") ~ {fe_terms}'

        try:
            # Ajustar modelo GLM con pesos de frecuencia
            model = smf.glm(
                formula=formula,
                data=df_collapsed,
                family=sm.families.Binomial(),
                freq_weights=df_collapsed["weight"]
            ).fit()

            # Calcular pseudo R¬≤ de McFadden manualmente
            ll_model = model.llf
            ll_null = model.null_deviance / -2
            pseudo_r2 = 1 - (ll_model / ll_null)

            # Guardar resultado
            pseudo_r2_results_sub_b.append({
                "Attribute": target,
                "FE_Set": label,
                "Pseudo_R2": pseudo_r2
            })

        except Exception as e:
            print(f"‚ö†Ô∏è Error with {target} / {label}: {e}")
            pseudo_r2_results_sub_b.append({
                "Attribute": target,
                "FE_Set": label,
                "Pseudo_R2": None,
                "Error": str(e)
            })

# Convertir resultados a DataFrame 
pseudo_r2_results_sub_b_df = pd.DataFrame(pseudo_r2_results_sub_b)
```
