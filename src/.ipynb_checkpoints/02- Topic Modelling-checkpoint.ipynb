## üìå Overview
An end-to-end Jupyter notebook that applies LDA topic modeling to a corpus of French job ads in order to surface novel, category-specific expressions (amenities) directly from the data. These discovered terms are used to enrich our attribute database, which will then drive scalable attribute scraping across the full dataset.  


## üì¶ Setup & Imports
Import core libraries (pandas, spaCy, Gensim, pyLDAvis), load the French NLP model, and define project file paths.


```python
# ============================================================================
# STEP 1: IMPORTS & PROJECT PATHS
# ============================================================================

# Standard Library
import csv
import os
import random
import re
import unicodedata
from collections import defaultdict
from pathlib import Path

# Data Manipulation
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis

# Natural Language Processing (NLP)
import nltk
from nltk.corpus import stopwords
import spacy
from bs4 import BeautifulSoup
from spacy.matcher import Matcher

# Topic Modeling
from gensim import corpora
from gensim.models import Phrases
from gensim.models.phrases import Phraser
from gensim.models.ldamodel import LdaModel
from gensim.models.coherencemodel import CoherenceModel

# Utilities
from unidecode import unidecode


# Define project directories
project_root = Path().resolve().parent
jocas_dir = "/Users/alfonso/Desktop/JOCAS"  # adjust if needed
file = project_root / "data" / "Attribute base" / "01- attribute_base_processed.xlsx"
```

## üì• Data Ingestion & Basic Filtering
Read all job-description CSVs into pandas, then apply length and frequency filters to remove outliers.


```python
# ============================================================================
# STEP 2: INGEST CSVs & BASIC FILTERING
# ============================================================================

# -------------------------------
# Initialize container
# -------------------------------
jocas_list = []

# -------------------------------
# Walk though source folders
# -------------------------------
for source_folder in os.listdir(jocas_dir):
    source_path = os.path.join(jocas_dir, source_folder)
    
    if os.path.isdir(source_path):
        # -------------------------------
        # Collect all csv files under this source
        # -------------------------------
        csv_files = []
        
        # Walk through all subdirectories (e.g., "January", "December", etc.)
        for root, dirs, files in os.walk(source_path):
            for file in files:
                # Case-insensitive check for .csv files
                if file.lower().endswith('.csv'):
                    csv_files.append(os.path.join(root, file))
        
        print(f"Found {len(csv_files)} CSV files in source folder: {source_folder}")
        
        if not csv_files:
            continue
        
        # -------------------------------
        # Select up to 50 csv files from this source
        # -------------------------------
        if len(csv_files) > 50:
            selected_files = random.sample(csv_files, 50)
        else:
            selected_files = csv_files
        
        print(f"Processing {len(selected_files)} files from source folder: {source_folder}")
        
        # Process each selected CSV file
        for file_path in selected_files:
            try:
                df = pd.read_csv(
                    file_path,
                    header=0,
                    sep=';',
                    on_bad_lines='skip',
                    low_memory=False,
                    quoting=csv.QUOTE_MINIMAL,
                    escapechar='\\'
                )
                jocas_list.append(df)
            except Exception as e:
                print(f"Failed to read {file_path}: {e}")

        print(f"Finished processing source folder: {source_folder}")

# -------------------------------
# Aggregate all dataframes into one
# -------------------------------
df_job = pd.concat(jocas_list, ignore_index=True)

# -------------------------------
# Flag English text and word counts
# -------------------------------
pd.set_option('display.max_colwidth', None)
def is_english(txt, threshold=2):
    common = ['the','and','for','with','that','this','from']
    return sum(bool(re.search(rf'\b{w}\b', txt, re.I)) for w in common) >= threshold

df_job['description_full'] = df_job['description_full'].astype(str)
df_job['word_count']      = df_job['description_full'].str.split().str.len()
df_job['is_english']      = df_job['description_full'].apply(is_english)
```

## ‚úÇÔ∏è Text Preprocessing for Topic Modeling
Remove French stop-words, tokenize and lemmatize text with spaCy, and clean noise tokens.


```python
# ============================================================================
# STEP 3: PREPROCESSING FOR TOPIC MODELLING
# ============================================================================

# ------------------------------------------------------------------------------
# SETUP
# ------------------------------------------------------------------------------
# Download NLTK stopwords if needed
nltk.download('stopwords')
french_stopwords = set(stopwords.words('french'))

# Load (or download) spaCy French model without parser & NER for speed
try:
    nlp = spacy.load("fr_core_news_sm", disable=["parser", "ner"])
except OSError:
    import spacy.cli
    spacy.cli.download("fr_core_news_sm")
    nlp = spacy.load("fr_core_news_sm", disable=["parser", "ner"])

# ------------------------------------------------------------------------------
# STEP 3.0: SAMPLE JOB DESCRIPTIONS
# ------------------------------------------------------------------------------
df_sample = (
    df_job
    .sample(n=100000, random_state=42)
    .copy()
)
df_sample['description_full'] = df_sample['description_full'].astype(str)
df_sample.dropna(subset=['description_full'], inplace=True)

# ------------------------------------------------------------------------------
# STEP 3.1: TEXT CLEANING & LEMMATIZATION
# ------------------------------------------------------------------------------
def clean_text(text, remove_accents=True):
    """Strip HTML, URLs, lowercase, remove accents, keep only letters/spaces."""
    # Remove HTML
    soup = BeautifulSoup(text, "html.parser")
    text = soup.get_text(separator=" ")
    # Drop URLs
    text = re.sub(r'http\S+|www\.\S+', ' ', text)
    text = text.lower()
    # Normalize accents
    if remove_accents:
        text = unicodedata.normalize('NFD', text)
        text = text.encode('ascii', 'ignore').decode('utf-8')
    # Keep only letters & spaces
    pattern = r'[^a-z\s]' if remove_accents else r'[^a-z√†√¢√ß√©√®√™√´√Æ√Ø√¥√ª√π√º√ø√±√¶≈ì\s]'
    text = re.sub(pattern, ' ', text)
    # Collapse whitespace
    return re.sub(r'\s+', ' ', text).strip()

def tokenize_lemmatize(text):
    """Return lemmas longer than 2 chars, alphabetic & not in French stopwords."""
    doc = nlp(text)
    return [
        tok.lemma_.lower()
        for tok in doc
        if tok.is_alpha
           and len(tok.lemma_) > 2
           and tok.lemma_.lower() not in french_stopwords
    ]

# Apply cleaning and tokenization
df_sample['clean'] = df_sample['description_full'].apply(lambda x: clean_text(x))
df_sample['tokens'] = df_sample['clean'].apply(tokenize_lemmatize)
df_sample = df_sample[df_sample['tokens'].map(len) > 0]

# ------------------------------------------------------------------------------
# STEP 3.2: BUILD BIGRAMS & TRIGRAMS
# ------------------------------------------------------------------------------
all_tokens = df_sample['tokens'].tolist()

# Train phrase models
bigram = Phrases(all_tokens, min_count=5, threshold=10)
trigram = Phrases(bigram[all_tokens], threshold=10)
bigram_mod = Phraser(bigram)
trigram_mod = Phraser(trigram)

def make_ngrams(tokens):
    """Merge tokens into bigrams & trigrams (underscore-joined)."""
    bigr = bigram_mod[tokens]
    trigr = trigram_mod[bigr]
    # Return only tokens that are true multi-word phrases
    return [t for t in bigr if '_' in t] + [t for t in trigr if '_' in t]

# Apply n-gram extraction and filter empties
df_sample['tokens_final'] = df_sample['tokens'].apply(make_ngrams)
df_sample = df_sample[df_sample['tokens_final'].map(len) > 0]

# ------------------------------------------------------------------------------
# STEP 3.3: DICTIONARY + CORPUS + FILTERING
# ------------------------------------------------------------------------------
# 3.3.1 Create Gensim Dictionary
dictionary = corpora.Dictionary(df_sample['tokens_final'])

# 3.3.2 Filter out rare and frequent tokens
dictionary.filter_extremes(
    no_below=5,      # keep tokens present in at least 5 docs
    no_above=0.5,    # keep tokens in at most 50% of docs
    keep_n=100000   # keep top 100k tokens by frequency
)

# 3.3.3 Build Corpus (Bag-of-Words for each doc)
corpus = [dictionary.doc2bow(doc) for doc in df_sample['tokens_final']]

# ------------------------------------------------------------------------------
# ‚îÄ‚îÄ‚îÄ SAVE FOR FUTURE USE ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# 1) Save the dictionary
# dictionary.save('job_dict.dict')  
#    ‚Üí writes a .dict file you can later load with corpora.Dictionary.load()
#
# 2) Serialize the corpus to Matrix Market format
# corpora.MmCorpus.serialize('job_corpus.mm', corpus)
#    ‚Üí writes a .mm file and an associated .mm.index file
# ------------------------------------------------------------------------------
```

## üîç Train LDA & Compute Coherence
Train LDA over a range of topic counts, calculate coherence scores, and plot to select the best model.


```python
# ------------------------------------------------------------------------------
# STEP 4: TRAIN LDA & COMPUTE COHERENCE
# ------------------------------------------------------------------------------

# ------------------------------------------------------------------------------
# STEP 4.1: TRAIN LDA WITH DIFFERENT TOPIC NUMBER + COHERENCE PLOT
# ------------------------------------------------------------------------------

def compute_coherence_values(
    dictionary,
    corpus,
    texts,
    start=5,
    limit=50,
    step=5,
    passes=15,
    alpha='auto'
):
    """
    Compute c_v coherence for various numbers of topics.

    Parameters:
    -----------
    dictionary : Gensim Dictionary object
    corpus     : Gensim Corpus (bag-of-words)
    texts      : List of token lists (e.g., df_sample['tokens_final'])
    start      : Minimum number of topics to try
    limit      : Maximum number of topics to try
    step       : Increment between topic counts
    passes     : Number of passes through the corpus during training
    alpha      : Dirichlet prior for document-topic distribution
    """
    model_list = []
    coherence_values = []

    for num_topics in range(start, limit + 1, step):
        # 4.1.1 Train LDA model
        model = LdaModel(
            corpus=corpus,
            id2word=dictionary,
            num_topics=num_topics,
            random_state=42,
            passes=passes,
            alpha=alpha,
            per_word_topics=False
        )
        model_list.append(model)

        # 4.1.2 Compute coherence
        coherencemodel = CoherenceModel(
            model=model,
            texts=texts,
            dictionary=dictionary,
            coherence='c_v'
        )
        coh = coherencemodel.get_coherence()
        coherence_values.append(coh)
        print(f"Trained LDA with {num_topics} topics ‚Äî coherence = {coh:.4f}")

    return model_list, coherence_values

# --- USAGE: explore coherence across topic counts ---
start, limit, step = 15, 25, 1
models, coherences = compute_coherence_values(
    dictionary=dictionary,
    corpus=corpus,
    texts=df_sample['tokens_final'],
    start=start,
    limit=limit,
    step=step,
    passes=15,
    alpha='auto'
)

# 4.1.3 Plot coherence scores
x = list(range(start, limit + 1, step))
plt.figure(figsize=(8, 4))
plt.plot(x, coherences, marker='o')
plt.xlabel("Number of Topics")
plt.ylabel("Coherence (c_v)")
plt.title("Coherence vs. Number of Topics")
plt.tight_layout()
plt.show()

# 4.1.4 Select best model
best_idx = coherences.index(max(coherences))
best_num_topics = x[best_idx]
best_model = models[best_idx]
print(f"\nOptimal number of topics: {best_num_topics} (c_v = {coherences[best_idx]:.4f})")



# ------------------------------------------------------------------------------
# STEP 4.2: TRAIN LDA WITH USER-DEFINED TOPIC COUNT
# ------------------------------------------------------------------------------
# Adjust these as needed:
num_topics = 19
passes     = 15
alpha      = 'auto'

lda_model = LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=num_topics,
    random_state=42,
    passes=passes,
    alpha=alpha,
    per_word_topics=True
)



# ------------------------------------------------------------------------------
# STEP 4.3: SAVE & VISUALIZE
# ------------------------------------------------------------------------------
# 4.3.1 Prepare visualization
vis = gensimvis.prepare(lda_model, corpus, dictionary)

# 4.3.2 Save to HTML
pyLDAvis.save_html(vis, f"lda_{num_topics}topics_vis.html")
print(f"Saved LDA model visualization: lda_{num_topics}topics_vis.html")
```

## üìä Build Topic‚ÄìTerm DataFrame
Extract the top terms for each topic into a DataFrame for manual inspection and downstream use.


```python
# ------------------------------------------------------------------------------
# STEP 5: BUILD DATAFRAME WITH MOST COMMON EXPRESSIONS IN TOPICS
# ------------------------------------------------------------------------------
records = []
for topic_id in range(num_topics):
    # lda_model.show_topic returns [(word, weight), ‚Ä¶] sorted by descending weight
    for rank, (word, weight) in enumerate(
            lda_model.show_topic(topic_id, topn=50),
            start=1
        ):
        records.append({
            'topic':  topic_id,
            'rank':   rank,
            'word':   word,
            'weight': weight
        })

# Create DataFrame of top-50 words per topic
df_top50 = pd.DataFrame(records)

# ‚îÄ‚îÄ‚îÄ SAVE FOR FUTURE USE ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# df_top50.to_csv('top50_words_per_topic_long.csv', index=False)

#--------------------------
# Looking at the dataframe manually, we extract expressions that
# appear as attributes (amenities...) for the next step
#--------------------------

# ------------------------------------------------------------------------------
# STEP 6: EXTRACT KEYWORD VARIANTS WITH spaCy
# ------------------------------------------------------------------------------
# 6.1 Load French model (disable unneeded components for speed)
nlp_matcher = spacy.load(
    "fr_core_news_sm",
    disable=["parser", "ner", "textcat"]
)

# 6.2 Define underscore-joined keywords
keywords = [
    "collaborateur_passionne", "ambiance_travail", "declaration_social",
    "plan_epargne", "etre_force_proposition", "force_proposition",
    "prendre_initiative", "total_autonomie", "sans_responsable",
    "autonomie_rigueur", "chef_projet", "grand_compte", "prise_decision",
    "prendre_charge", "voir_confier", "permettre_epanouir", "haut_niveau",
    "ete_passionn", "innovation_passion", "deplacement_ponctuel",
    "deplacement_occasionnel", "mission_variee", "bien_etre",
    "contrat_cdi", "cdd_cdi", "temps_partiel", "temps_complet",
    "temps_plein", "week_end", "fort_croissance", "opportunite_carriere",
    "faire_progresser", "montee_grade", "possibilite_evoluer",
    "evolution_carriere", "montee_competence", "amelioration_continuer",
    "devenir_expert", "conseil_carriere", "formation_tout_long",
    "brut_annuel", "remuneration_brut", "euro_brut", "remuneration_avantage",
    "remuneration_attractif", "convention_collectif", "salair_fixe",
    "pourvoir_possible_remuneration", "remuneration_selon", "salair_selon",
    "mois_prime", "taux_horaire", "minimum_garanti", "souhait_prime",
    "prime_collectif", "salaire_brut", "brut_mensuel", "annuel_brut",
    "maison_individuel", "ticket_restaurant", "frais_deplacement",
    "cheque_vacance", "conge_paye",
]

# 6.3 Initialize the Matcher
matcher = Matcher(nlp_matcher.vocab)
max_span = 6  # maximum tokens in a matched span

for kw in keywords:
    parts = kw.split("_")
    if len(parts) == 1:
        # single-lemma keyword
        pattern = [{"LEMMA": parts[0]}]
    else:
        # allow any alpha tokens between first and last lemma
        pattern = [
            {"LEMMA": parts[0]},
            {"IS_ALPHA": True, "OP": "*"},
            {"LEMMA": parts[-1]}
        ]
    matcher.add(kw, [pattern])

# 6.4 Run matcher over all job descriptions
variants = defaultdict(set)
for doc in nlp_matcher.pipe(df_sample['description_full'].tolist(),
                             batch_size=32, n_process=2):
    for match_id, start, end in matcher(doc):
        if end - start <= max_span:
            key = nlp_matcher.vocab.strings[match_id]
            variants[key].add(doc[start:end].text)

# 6.5 Convert sets to lists and display
variants = {k: list(v) for k, v in variants.items()}
for kw, forms in variants.items():
    print(f"\n=== {kw} ({len(forms)} variants) ===")
    for form in forms:
        print(" ‚Ä¢", form)
```

## üõ†Ô∏è Extend Stop-Words & Extract Lemma Variants
Enhance the French stop-word list, then generate and inspect lemma variant mappings to refine text cleaning.


```python
# ------------------------------------------------------------------------------
# STEP 7.1: EXTEND STOP-WORDS & EXTRACT LEMMA VARIANTS
# ------------------------------------------------------------------------------

# 7.1.1 Download NLTK French stop-words and build base set
nltk.download('stopwords')
french_stopwords = set(stopwords.words('french'))

# 7.1.2 Define extra ‚Äúin-between‚Äù tokens to filter out
additional_stopwords = [
    'alors', 'assez', 'avoir', 'bien', 'cahier', 'connaissance',
    'directement', 'essentiellement', 'ensuite', 'egalemer',
    'egalement', 'integralement', 'notamment', 'parfaitement',
    'particulieremer', 'plus', 'principalement', 'prioritairement',
    'progressivement', 'progressivemer', 'rapidement',
    'aussi', 'ainsi', 'donc', 'constant', 'continues', 'prevue',
    'prevues', 'possibles', 'possible', 'sont', 'individualisee',
    'soutien', 'personnalise', 'continue', 'veritable', 'tres',
    'tre', 'veritabl', 'extremement', 'futur'
]

# 7.1.3 Merge base and additional stop-words
french_stopwords |= set(additional_stopwords)

# 7.1.4 Load spaCy French model (disable parser, NER, textcat for speed)
nlp_stop = spacy.load("fr_core_news_sm", disable=["parser", "ner", "textcat"])

# ------------------------------------------------------------------------------
# STEP 7.2: PREPROCESSING FUNCTION
# ------------------------------------------------------------------------------
def preprocess_text(text: str) -> str:
    """
    Lowercase, strip accents, tokenize & lemmatize, remove stop-words.
    Returns a space-joined string of lemmas.
    """
    # Lowercase and remove accents
    txt = unidecode(text).lower()

    # Lemmatize and filter out stop-words
    doc = nlp_stop(txt)
    lemmas = [
        tok.lemma_.lower()
        for tok in doc
        if tok.is_alpha and tok.lemma_.lower() not in french_stopwords
    ]
    return " ".join(lemmas)

# ------------------------------------------------------------------------------
# STEP 7.3: BUILD LEMMA VARIANT MAPPING
# ------------------------------------------------------------------------------
lemma_variants = defaultdict(set)

for kw, forms in variants.items():
    for form in forms:
        lemma_form = preprocess_text(form)
        if lemma_form:  # skip empty results
            lemma_variants[kw].add(lemma_form)

# Convert each set of lemma-forms to a list
lemma_variants = {kw: list(s) for kw, s in lemma_variants.items()}

# ------------------------------------------------------------------------------
# STEP 7.4: INSPECT UNIQUE LEMMA FORMS
# ------------------------------------------------------------------------------
for kw, lemmas in lemma_variants.items():
    print(f"\n=== {kw} ({len(lemmas)} unique lemma forms) ===")
    for lm in lemmas:
        print(" ‚Ä¢", lm)
```

```python
# ------------------------------------------------------------------------------
# STEP 8: UPDATE ATTRIBUTE BASE WITH NEW EXPRESSIONS
# ------------------------------------------------------------------------------

# 8.1 Load your existing attribute table
# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
attribute_base = pd.read_excel(file)

# 8.2 Define new attributes and their expressions
# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
new_entries = {
    "Tickets restaurants": [
        "ticket restau",
        "ticket resto",
    ],
    "Ch√®ques vacance": [
        "cheque vacance",
    ],
    "Responsabilit√©s dans le poste": [
        "prendra en charge",
        "prendre en charge",
        "prendrez en charge",
        "grand compte",
        "voir confier",
        "verrez confier",
        "prise de decision",
    ],
    "Autonomie dans l‚Äôex√©cution des t√¢ches": [
        "sans responsable",
        "force proposition",
        "autonomie",
    ],
    "Niveau de r√©mun√©ration": [
        "taux horaire",
        "taux brut horaire",
        "euro",
        "‚Ç¨",
    ],
    "Environnement social": [
        "reconnaissance travail",
        "ambiance travail",
        "ambiance agreabl",
        "ambiance favorable",
        "ambiance stimulant",
        "collaborateurs passionnes",
    ],
    "Bonnes perspectives de carri√®re": [
        "fort potentiel de croissance",
        "fortes perspectives de croissance",
        "forte politique de croissance",
        "evolution de carriere",
        "opportunites de carriere",
    ],
    "Responsabilit√© de leadership incluse": [
        "chef de projet",
    ],
    "Toute autre mention de salaire": [
        "minimum r√©mun√©r√© garanti",
        "minimum indicatif est garanti",
        "minimum mensuel garanti",
    ],
    "Contrat √† temps plein": [
        "temps complet",
    ],
    "Choix entre temps plein et temps partiel": [
        "temps partiel complet",
        "temps partiel possibilite temp complet",
        "temps partiel temps complet",
        "temps partiel temps plein",
        "temps partiel plein",
    ],
    "Formation en cours d‚Äôemploi": [
        "formation professionnelle continue",
        "formation tout au long",
        "Formation assur√©e",
        "formations au long",
        "devenir expert",
        "amelioration continuer",
        "montee en competences",
    ],
    "Salaire comp√©titif": [
        "remuneration attractive",
        "remuneration individuelle et collective attractive",
        "remuneration interessant",
    ],
    "Vari√©t√© des t√¢ches": [
        "mission riche variees",
    ],
}

# 8.3 Manually specify aggregates for new attributes category
# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
aggregate_mapping = {
    "Tickets restaurants": "Avantages mineurs",
    "Ch√®ques vacance":   "Avantages mineurs",
}

# 8.4 Fill in aggregate mapping for existing attributes
# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
for attr in new_entries:
    if attr not in aggregate_mapping:
        existing = attribute_base.loc[
            attribute_base["attribute_category_fr"] == attr,
            "aggregate_category_fr"
        ].unique()
        if len(existing) == 0:
            raise ValueError(f"No aggregate_category_fr found for attribute ‚Äú{attr}‚Äù")
        aggregate_mapping[attr] = existing[0]

# 8.5 Build list of new rows (if not already present)
# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
rows_to_add = []
for attr, expr_list in new_entries.items():
    agg = aggregate_mapping[attr]
    for expr in expr_list:
        mask = (
            (attribute_base["attribute_category_fr"] == attr) &
            (attribute_base["expression"] == expr)
        )
        if not mask.any():
            rows_to_add.append({
                "aggregate_category_fr": agg,
                "attribute_category_fr": attr,
                "expression":            expr,
            })

# 8.6 Append new rows and save updated table
# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
# Append
attribute_base = pd.concat(
    [attribute_base, pd.DataFrame(rows_to_add)],
    ignore_index=True
)
attribute_base.sort_values("attribute_category_fr", inplace=True)

# Save
output_file = Path(project_root) / "data" / "attribute_base_processedV2.xlsx"
attribute_base.to_excel(output_file, index=False)

# Report
print(f"Added {len(rows_to_add)} new rows and wrote out updated Excel.")
```
